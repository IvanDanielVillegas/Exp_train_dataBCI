```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import signal
import math
import mne
%matplotlib inline
```

```python
epochs = mne.read_epochs('/content/drive/MyDrive/BCI/BCICompetition/data/oddball_test_epo.fif')
```

```python
target = epochs["P300"]
target
```

```python
standart = epochs["NoP300"].average()
standart
```

```python
epochs.plot_psd(fmin=0,fmax=40)
```

```python
from mne.time_frequency import tfr_morlet
```

```python
mne.time_frequency.tfr_morlet?
```

```python
freqs = list(range(3,40))
tfr_target = tfr_morlet(target ,freqs ,1,return_itc=False)
tfr_standard = tfr_morlet(standart,freqs,1,return_itc=False)
```

```python
print(tfr_target.data.shape)
tfr_target.plot()
```

```python
tfr_contrast = mne.combine_evoked((tfr_standard, tfr_target),(-.5,.5))
tfr_contrast.apply_baseline((None,0))
```

```python
tfr_contrast.plot_joint()
```

```python
tfr_contrast.plot(picks=[11])
```
